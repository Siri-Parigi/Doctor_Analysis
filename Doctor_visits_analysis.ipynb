# Visualize the health issue by using pie chart

```python
data = pd.read_csv('DoctorVisit.csv')
class_counts = data['SMOKING'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of SMOKING Classes')
plt.axis('equal')
plt.show()
```

# Plot a horizontal bar chart to analyse the  Anxiety based on age

```python
db = df.groupby('AGE')['ANXIETY'].sum().to_frame().reset_index()
plt.barh(db['AGE'],db['ANXIETY'],color =['cornflowerblue','lightseagreen'] )
plt.title('Bar Chart')
plt.xlabel('AGE')
plt.ylabel(' ANXIETY')
plt.show()
```
